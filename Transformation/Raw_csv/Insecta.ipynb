```python
df.RF.isna().sum()
```

```python
# Dropping RF Column
df.drop(columns = "RF", inplace = True)
```

```python
df.head()
```

```python
df
```

```python
# Editing species column based on binomial nomenclature

df1 = df.Species.str.split(pat=" ", n = 3, expand = True)
df.Species = df1[0].str.cat(df1[1], sep = " ")
```

```python
df
```

```python
# Manually editing cells that were extracted incorrectly/partially
df.at[0,"GfU"]= "1c, 2d, 3b, 11c, 14a, 14g"
df.head()
```

```python
df.at[2,"GfU"]= "1c, 2d, 8b, 11c, 14a"
df.at[3,"GfU"]= "1c, 2d, 11c, 14a, 14g"
df.at[4,"GfU"]= "1c, 2d, 11c, 14a, 14g"
df
```

```python
df.at[24,"GfU"]= "2d, 5b, 6a, 14a, 14g"
df.at[28,"GfU"]= "2d, 8b, 11c, 14a, 14g"
```

```python
df.at[8,"Quelle"]= "coll. G. Lauterbach in FU Berlin"
df.at[16,"Quelle"]= "coll. J. Neresheimer, MOHR (1977)"
df.at[21,"Quelle"]= "coll. G. Kraatz (SDEI), Mohr & Arnold in litteris"
df.at[24,"Quelle"]= "NERESHEIMER & WAGNER (1917)"
df.at[30,"Quelle"]= "WENDT (1986), CHRYFAUN, SMF"
```

```python
df
```

```python
# Adding aditional information columns from the source
```

```python
# Adding the Family and SubFamily columns
```

```python
#df = df.reindex(columns = ["Family", "Sub_family"] + df.columns.tolist())
```

```python
i = 0;
while i < 32:
    df.at[i,"Family"] = "Chrysomelidae"
    if i < 6:
        df.at[i,"Sub_family"] = "Donaciinae"
    elif i == 6:
        df.at[i,"Sub_family"] = "Criocerinae"
    elif (i == 7) | (i == 8):
        df.at[i,"Sub_family"] = "Cryptocephalinae – Clytrini" 
    elif ((i > 8) & (i < 21)):
        df.at[i,"Sub_family"] = "Cryptocephalinae – Cryptocephalini"
    elif (i == 21):
        df.at[i,"Sub_family"] = "Lamprosomatinae"
    elif ((i > 21) & (i < 25)):
        df.at[i,"Sub_family"] = "Chrysomelinae"
    elif (i == 25):
        df.at[i,"Sub_family"] = "Galerucinae – Galerucini"
    elif ((i > 25) & (i < 28)):
        df.at[i,"Sub_family"] = "Galerucinae – Alticini"
    elif ((i > 27)):
        df.at[i,"Sub_family"] = "Bruchinae"
    i = i + 1
df
```

```python
df.columns
```

```python
df = df.reindex(columns=['Family',
                         'Sub_family',
                         'Species', 
                         'BE', 
                         'Bestand', 
                         'BB', 
                         'D', 
                         'GfU', 
                         'Letzter_Nachweis', 
                         'Quelle'])
df
```

```python
# Removing column Quelle, GfU, BB, D
df.drop(columns = ["Quelle", "GfU", "BB", "D"], inplace = True)
```

```python
df
```

```python
# Saving the extracted csv to the "Data_cleaning_step1" directory
# export the dataframe as CSV

df.to_csv('../../Transformation/Data_cleaning_step1/Chrysomelidae_ Megalopodidae_Blattkäfer.csv', index=False)
```

# 03.  Coleoptera_Wasserkäfer.csv

```python
# Read file into dataframe

df = pd.read_csv('Coleoptera_Wasserkäfer.csv')
df.head()
```

```python
df
```

```python
# Adding family

i = 0;
while i < 31:
    if i == 0:
        df.at[i,"Family"] = "Hygrobiidae"
    elif i == 1:
        df.at[i,"Family"] = "Haliplidae"
    elif ((i > 1) & (i < 21)):
        df.at[i,"Family"] = "Dytiscidae" 
    elif ((i > 20) & (i < 25)):
        df.at[i,"Family"] = "Gyrinidae"
    elif (i == 25):
        df.at[i,"Family"] = "Hydrochidae"
    elif (i == 26):
        df.at[i,"Family"] = "Helophoridae"
    elif (i == 27):
        df.at[i,"Family"] = "Elmidae"
    else:
        df.at[i,"Family"] = "Dryopidae"
    i = i + 1

df
```

```python
df = df.reindex(columns=['Family',
                         'Species', 
                         'RL', 
                         'Letzter Nachweis in Berlin'])
df
```

```python
df.rename({
    "RL": "BE",
    "Letzter Nachweis in Berlin": "Letzter Nachweis"
}, 
    axis='columns',
    inplace = True
)
```

```python
df.head()
```

```python
# Saving the extracted csv to the "Data_cleaning_step1" directory
# export the dataframe as CSV

df.to_csv('../../Transformation/Data_cleaning_step1/Coleoptera_Wasserkäfer.csv', 
          index=False)
```
