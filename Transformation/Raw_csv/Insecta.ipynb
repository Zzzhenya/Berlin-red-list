```python
# Adding family

i = 0;
while i < 31:
    if i == 0:
        df.at[i,"Family"] = "Hygrobiidae"
    elif i == 1:
        df.at[i,"Family"] = "Haliplidae"
    elif ((i > 1) & (i < 21)):
        df.at[i,"Family"] = "Dytiscidae" 
    elif ((i > 20) & (i < 25)):
        df.at[i,"Family"] = "Gyrinidae"
    elif (i == 25):
        df.at[i,"Family"] = "Hydrochidae"
    elif (i == 26):
        df.at[i,"Family"] = "Helophoridae"
    elif (i == 27):
        df.at[i,"Family"] = "Elmidae"
    else:
        df.at[i,"Family"] = "Dryopidae"
    i = i + 1

df
```

```python
df = df.reindex(columns=['Family',
                         'Species', 
                         'RL', 
                         'Letzter Nachweis in Berlin'])
df
```

```python
df.rename({
    "RL": "BE",
    "Letzter Nachweis in Berlin": "Letzter Nachweis"
}, 
    axis='columns',
    inplace = True
)
```

```python
df.head()
```

```python
# Saving the extracted csv to the "Data_cleaning_step1" directory
# export the dataframe as CSV

df.to_csv('../../Transformation/Data_cleaning_step1/Coleoptera_Wasserkäfer.csv', 
          index=False)
```

# 04.Coleoptera_Bostrichoidea_Cleroidea_Cucujoidea_Elateroidea_Lymexyloidea_Tenebrioidea.csv

```python
# Read file into dataframe

df = pd.read_csv('Coleoptera_Bostrichoidea_Cleroidea_Cucujoidea_Elateroidea_Lymexyloidea_Tenebrioidea.csv')
df.head()
```

```python
df.describe()
```

```python
# Removing columns Trend_lang, Trend_kurz, RF, BB, D, GS, Neobiota
df.drop(columns = ["Trend_lang", "Trend_kurz", "RF", "BB", "D", "GS", "Neobiota", "Bestand"], inplace = True)
df
```

```python
# Editing species column based on binomial nomenclature

df1 = df.Species.str.split(pat=" ", n = 3, expand = True)
df1
df.Species = df1[0].str.cat(df1[1], sep = " ")
```

```python
df.head()
```

```python
# Adding family

i = 0;
while i < 40:
    if (i >= 0) & (i <= 4):
        df.at[i,"Superfamily"] = "Bostrichoidea"
        
        if i == 0:
            df.at[i,"Family"] = "Bostrichidae"
        elif i == 1:
            df.at[i,"Family"] = "Endecatomidae"
        elif ((i > 1) & (i <= 4)):
            df.at[i,"Family"] = "Ptinidae" 
    elif (i > 4) & (i <= 11):
        df.at[i,"Superfamily"] = "Cleroidea"
        
        if ((i > 4) & (i <= 8)):
            df.at[i,"Family"] = "Cleridae"
        elif (i == 9):
            df.at[i,"Family"] = "Dasytidae"
        elif (i == 10):
            df.at[i,"Family"] = "Phloiophilidae"
        elif (i == 11):
            df.at[i,"Family"] = "Trogossitidae"
            
    elif (i > 11) & (i <= 24):
        df.at[i,"Superfamily"] = "Cucujoidea"
        
        if ((i > 11) & (i <= 17)):
            df.at[i,"Family"] = "Cleridae"
        elif (i == 18):
            df.at[i,"Family"] = "Cucujidae"
        elif ((i > 18) & (i <= 20)):
            df.at[i,"Family"] = "Erotylidae"
        elif ((i > 20) & (i <= 22)):
            df.at[i,"Family"] = "Laemophloeidae"
        elif (i == 23):
            df.at[i,"Family"] = "Monotomidae"
        elif (i == 24):
            df.at[i,"Family"] = "Nitidulidae"
            
    elif (i > 24) & (i <= 29):
        df.at[i,"Superfamily"] = "Elateroidea"

        if ((i > 24) & (i <= 28)):
            df.at[i,"Family"] = "Elateridae"
        elif (i == 29):
            df.at[i,"Family"] = "Throscidae"           
            
    elif (i > 29):
        df.at[i,"Superfamily"] = "Tenebrioidea"
        
        if (i == 30):
            df.at[i,"Family"] = "Colydiidae"
        elif ((i > 30) & (i <= 32)):
            df.at[i,"Family"] = "Melandryidae"
        elif (i == 33):
            df.at[i,"Family"] = "Oedemeridae"
        elif (i == 34):
            df.at[i,"Family"] = "Salpingidae"
        else:
            df.at[i,"Family"] = "Tenebrionidae"
                
    i = i + 1

df
```

```python
# Saving the extracted csv to the "Data_cleaning_step1" directory
# export the dataframe as CSV

df.to_csv('../../Transformation/Data_cleaning_step1/Coleoptera_Bostrichoidea_Cleroidea_Cucujoidea_Elateroidea_Lymexyloidea_Tenebrioidea.csv', 
          index=False)
```
