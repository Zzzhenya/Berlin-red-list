```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import seaborn as sns
```

```python
df = pd.read_csv('Butterflies_and_moths_final_updated - v2.csv')
df.head(5)
```

```python
df.describe()
```

```python
counts_by_family = df.groupby(['Familie'])[['Familie']].count()
counts_by_family.rename(columns={"Familie": "Count"}, inplace=True)
counts_by_family.reset_index(inplace = True)
counts_by_family
```

```python
ax = sns.barplot(data=counts_by_family, 
            x = "Familie", 
            y = "Count", 
            order = counts_by_family.sort_values('Count',ascending = False).Familie,
            palette = "RdBu"
           )
sns.set(rc={'figure.figsize':(20,9)}, 
        style = "whitegrid"
       )

ax.axes.set_title("Number of extinct butterfly and moth species by family",fontsize=25)
for i in ax.containers:
    ax.bar_label(i,)
```

```python
counts_by_g = df.groupby(['Genus'])[['Genus']].count()
counts_by_g.rename(columns={"Genus": "Count"}, inplace=True)
counts_by_g.reset_index(inplace = True)
counts_by_g
```

```python
ax = sns.barplot(data=counts_by_g, 
            x = "Count", 
            y = "Genus", 
            order = counts_by_g.sort_values('Count',ascending = False).Genus,
            palette = "RdBu"
           )
sns.set(rc={'figure.figsize':(20,20)}, 
        style = "whitegrid"
       )

ax.axes.set_title("Number of extinct butterfly and moth species by Genus",fontsize=25)
for i in ax.containers:
    ax.bar_label(i,)
```
