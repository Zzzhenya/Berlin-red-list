# Concating

```python
# Creates a sorted dictionary (sorted by key)
from collections import OrderedDict
d2 = OrderedDict(sorted(d.items()))
```

```python
for key in d2:
    print(key)
```

```python
for key in d2:
    print(key)
    d2[key].columns = [ "Species",
    "BE", 
    "Bestand", 
    "Vor_1990", 
    "1990_2017",
    "Trend_kurz",
    "Habitat",
    "D",
    "BB",
    "GS",
    "Common_name"]
print("Rename completed for concat")
```

```python
for key in d2:
    third_61_89 = pd.concat([third_61_89, d2[key]], axis = 0)
third_61_89
```
